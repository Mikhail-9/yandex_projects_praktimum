### Сгрупируем добавленные данные.

```python
counted_metricks = calls_grouped.merge(internet_grouped, on=['user_id', 'month'], how='outer').merge(messages_grouped, on=['user_id', 'month'], how='outer')
counted_metricks
```

###### Объеденим users и counted_metricks

```python
users_merged = users.merge(counted_metricks, on='user_id', how='right')
display(users_merged)
```

###### Переименуем столбцы с тарифами единообразно, для объединения таблиц. 

```python
tariffs.columns = ['messages_included', 'mb_per_month_included', 'minutes_included', 'rub_monthly_fee', 'rub_per_gb', 'rub_per_message', 'rub_per_minute', 'tariff']
```

```python
tariffs
```

###### Объеденим сводную таблицу с тарифами, приведем столбец 'mb_per_month_included' к гигабайтам.

```python
users_merged = users_merged.merge(tariffs, on='tariff', how='right')
users_merged['mb_per_month_included'] = (users_merged['mb_per_month_included'] / 1024).astype(int)
display(users_merged)
```

###### Напишем функцию дохода для одной строки

```python
def revenue(row):
    charge = 0
    if row['duration_sum'] > row['minutes_included']:
        add_duration = (row['duration_sum'] - row['minutes_included']) * row['rub_per_minute']
        charge += add_duration
    else:
        charge += 0
    if row['msg_sum'] > row['messages_included']:
        add_msg = (row['msg_sum'] - row['messages_included']) * row['rub_per_message']
        charge += add_msg
    else:
        charge += 0
    if row['session_sum'] > row['mb_per_month_included']:
        add_session = (row['session_sum'] - row['mb_per_month_included']) * row['rub_per_gb']
        charge += add_session
    else:
        charge += 0
    charge += row['rub_monthly_fee']
    return charge       
            
users_merged['revenue'] = users_merged.apply(revenue, axis=1)    
users_merged
```

## Проверка гипотез

### Средняя выручка пользователей тарифов "Ультра" и "Смарт" различаются

* Это гипотеза о равенстве среднего двух генеральных совокупностей по взятым из них выборкам 

#### Нулевая гипотеза Hо - средняя выручка тарифов "Ультра" и "Смарт" равны. Альтернативная гипотеза H1 - средняя выручка тарифов "Ультра" и "Смарт" различаются.

```python
ultra = users_merged.query('tariff == "ultra"')['revenue']
smart = users_merged.query('tariff == "smart"')['revenue']

alpha = .01

results = st.ttest_ind(ultra, smart, equal_var=False)

display('p-значение', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

### Средняя выручка пользователей из Москвы отличается от выручки пользователей других регионов.
