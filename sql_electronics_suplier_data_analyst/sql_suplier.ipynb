**Задание: Вычислить среднюю цену ПК & принтеров & пк-блокнотов, которые производят компании A или D (на английском). 
Вывести: одна средняя цена.**

```python
request_task = """
SELECT AVG(total_base.total_price)
FROM(
SELECT p.maker,
       pc.price AS total_price
FROM product AS p
JOIN pc AS pc ON p.model=pc.model
WHERE p.maker = 'A' 
   OR p.maker = 'D'
UNION ALL
SELECT p.maker,
       l.price AS total_price
FROM product AS p
JOIN laptop AS l ON p.model=l.model
WHERE p.maker = 'A' 
   OR p.maker = 'D'
UNION ALL
SELECT p.maker,
       pr.price AS total_price
FROM product AS p
JOIN printer AS pr ON p.model=pr.model
WHERE p.maker = 'A' 
   OR p.maker = 'D') AS total_base

     
     

"""

c.execute(request_task)
results = c.fetchall()
results
```
