**Небольшой обзор, пропусков нет(на первый взгляд:), две последних таблицы содержат меньше данных чем первые, причина в том что клиенты могут пользоваться разными услугами независимо - заполнить пропуски. Не соответсвтие типов данных и столбцов в таблице contract 'BeginDate', 'EndDate' - привести к дате, 'TotalCharges' к float.В таблице personal столбец 'SeniorSitizen' числовой хотя это категоральный признак. Разный регистр, нет целевого признака.**

**Объеденим таблицы в один датасет, приведем названия столбцов к одному регистру**

```python
df = data['contract'].join(data['personal']).join(data['internet']).join(data['phone'])
df.columns = [k.lower() for k in list(df.columns)]
display(df.sample(5))
df.info()
```

**Создадим столбец 'outflow' с целевым признаком**

```python
def func(row):
    if row.enddate == "No":
        return 0
    else:
        return 1
df['outflow'] = df.apply(func, axis=1) 
```

```python
df
```

**Заменим столбцы 'begindate', 'enddate' на столбец 'contract_length' так мы избавимся от утечки целевого признака и сохраним информативность временного признака. Для начала приведём их к дате и заменим пропуски на дату предоставления данных**

```python
df['begindate'] = pd.to_datetime(df['begindate'], errors='coerce')
df['enddate'] = pd.to_datetime(df['enddate'], errors='coerce')
```

```python
df['enddate'].fillna(pd.to_datetime('2020-02-01'), inplace=True)
```

```python
def duration(row):
    return len(pd.date_range(start=row['begindate'], end=row['enddate']))
df['contract_length'] = df.apply(duration, axis=1)
df.drop(['begindate', 'enddate'], axis=1, inplace=True)
```

```python
df
```

**Приведём данные столбца totalcharges к вещественному типу.**

```python
df['totalcharges'] = pd.to_numeric(df['totalcharges'], errors='coerce')
```

**Изучим пропуски**

```python
df[df['totalcharges'].isna()]
```

**Пропуски связаны с дилиноой контракта в один день, значит можем заменить их на 0**

```python
df['totalcharges'].fillna(0, inplace=True)
```

**Приведём столбец 'seniorsitezen' к категоральному виду, каким он и является, заменив '1', '0' на ''YES, 'NO'.**

```python
df['seniorcitizen'] = np.where(df['seniorcitizen'] == 1, 'YES', 'NO')
```

**Проверка**

```python
df
```

**Изучим уникальные значения категоральных прризнаков**

```python
cat_features = np.array(df.select_dtypes(include=['object']).columns)
for z in cat_features:
    print(z + ':', df[z].unique())
```

**Заполним пропуски 'None'**

```python
df.fillna('None', inplace=True)
display(df.sample(7))
df.info()
```

**Проверим на дубли**

```python
df.duplicated().sum()
```

_Удалим дубли_

```python
df = df.drop_duplicates().reset_index(drop=True)
```

```python
df.duplicated().sum()
```

**Для открытия результатов используйте jupiter ==> file\open...\outflow.html**

**Выводы:**

* Уровень изменения расчитан относительно одного признака

* type – тип оплаты: ежемесячный, годовой двухлетний, отток клиентов максимален при помесячной оплате в четыре раза выше 
  чем при годовой и в десять чем при двухлетнем контракте
  
* paperlessbilling – безбумажное выставление счетов, стимулирует к уходу в два раза чаще, по сравнению с бумажным

* paymentmethod – способ оплаты, электронный чек в два раза повышает отток по сравнению с чеком по почте, бакновским             переводом(автоматическим), списанием с кредитной карты(автоматическим)

* monthlycharges – ежемесячные траты на услуги, показывают очень слабую корреляцию при минимальном значении зависимости в точке   60

* totalcharges – всего потрачено денег на услуги показывают очень слабую корреляцию при тенденции снижения при увеличении суммы 
  расходов
  
* gender - пол, выявлен паритет

* senior Citizen – наличие пенсионного статуса по возрасту, пенсионеры склонны в два раза чаще отказываться от услуг

* partner – наличие супруга(и)  влияет слегка положительно и снижает отток на по этому признаку на треть, при очень слабой       корреляции

* dependents – наличие иждивенцев, улучшает отток(снижает) в два раза

* internetservise - наличие  оптики увеличивает отток в два раза по сравнению с DSL

* onlinesecurity - наличие сервиса снижает отток почти  втри раза

* onlinebackup - резеврное копирование онлайн снижает отток в два раза

* deviceprotection - защита устройства удучшает отток на 42 %

* techsupport - техническая поддержка почти в три раза уменьшает отток

* streamingtv - стриминговое тв не влияет на отток

* streamingmovies - стриминговый кинотеатр не влияет на отток

* multipleLines – наличие возможности ведения параллельных линий во время звонка не влияет на отток

* contract_length - длина контракта максимальное количество расторжений в первые дни и затем снижается
