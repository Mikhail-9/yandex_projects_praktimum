**Объеденим таблицы в один датасет, приведем названия столбцов к одному регистру**

```python
df = data['contract'].join(data['personal']).join(data['internet']).join(data['phone'])
df.columns = [k.lower() for k in list(df.columns)]
display(df.sample(5))
df.info()
```

**Создадим столбец 'outflow' с целевым признаком**

```python
def func(row):
    if row.enddate == "No":
        return 0
    else:
        return 1
df['outflow'] = df.apply(func, axis=1) 
```

```python
df
```

**Заменим столбцы 'begindate', 'enddate' на столбец 'contract_length' так мы избавимся от утечки целевого признака и сохраним информативность временного признака. Для начала приведём их к дате и заменим пропуски на дату предоставления данных**

```python
df['begindate'] = pd.to_datetime(df['begindate'], errors='coerce')
df['enddate'] = pd.to_datetime(df['enddate'], errors='coerce')
```

```python
df['enddate'].fillna(pd.to_datetime('2020-02-01'), inplace=True)
```

```python
def duration(row):
    return len(pd.date_range(start=row['begindate'], end=row['enddate']))
df['contract_length'] = df.apply(duration, axis=1)
df.drop(['begindate', 'enddate'], axis=1, inplace=True)
```

```python
df
```

**Приведём данные столбца totalcharges к вещественному типу.**

```python
df['totalcharges'] = pd.to_numeric(df['totalcharges'], errors='coerce')
```

**Изучим пропуски**

```python
df[df['totalcharges'].isna()]
```

**Пропуски связаны с дилиноой контракта в один день, значит можем заменить их на 0**

```python
df['totalcharges'].fillna(0, inplace=True)
```

**Приведём столбец 'seniorsitezen' к категоральному виду, каким он и является, заменив '1', '0' на ''YES, 'NO'.**

```python
df['seniorcitizen'] = np.where(df['seniorcitizen'] == 1, 'YES', 'NO')
```

**Проверка**

```python
df
```

```python
cat_features = np.array(df.select_dtypes(include=['object']).columns)
for z in cat_features:
    print(z + ':', df[z].unique())
```

__Признак totalcharges сильно коррелирует с признаками contract_length и monthlycharges, удалим его , чтобы избежать мультиколлинеарности.__

**Дополнительные сервисы оказывают влияние на отток,т.к. они в отдельности каждый уменьшают отток, то чем их больше тем меньше отток, создадим признак показывающий количество сервисов у клиента, заменив им все дополнительные сервисы удалив их. Также удалим gendre как неинформативный, и сбросим индексы**

```python
additional_services = df.iloc[:, 10:16].columns

def services(row):
    n=0
    for i in additional_services:
        if row[i]=='Yes':
            n+=1
        if row[i]=='None':
            n=-1
    return n

df['services'] = df.apply(services, axis=1)
```

```python
df.reset_index(drop=True, inplace=True)
df.drop(list(additional_services) + ['gender', 'totalcharges'], axis=1, inplace=True)
```

## Построение модели

### CatBoost

**Создадим списки признаков**

```python
X = ['monthlycharges', 'contract_length', 'type', 'paperlessbilling', 'paymentmethod', 'seniorcitizen',
 'partner', 'dependents', 'internetservice', 'services',
'multiplelines']

cat_features = ['type', 'paperlessbilling', 'paymentmethod', 'seniorcitizen',
 'partner', 'dependents', 'internetservice', 'multiplelines'] 

y = ['outflow']
```

**Создадим обучающую и тестовую выборку**

```python
train, test = train_test_split(df,train_size=0.8,random_state=77)
```

**Создание POOL**

```python
train_data = Pool(data=train[X],
                  label=train[y],
                  cat_features=cat_features
                 )
```

```python
test_data = Pool(data=test[X],
                  label=test[y],
                  cat_features=cat_features
                 )
```

**Кросс валидация**

```python
params = {'cat_features': cat_features,
          'eval_metric': 'AUC',
          'loss_function': 'Logloss',
          'learning_rate': 0.11,
          'random_seed': 77,
          'verbose':100,
          'scale_pos_weight': 74/26}
```

```python
cv_data = cv(
    params = params,
    pool = train_data,
    fold_count=5,
    shuffle=True,
    partition_random_seed=0,
    plot=True,
    stratified=True,
    verbose=False,
    early_stopping_rounds=200
)
```

**Найдём лучшую итерацию**

```python
cv_data.head()
```

```python
cv_data[cv_data['test-AUC-mean'] == cv_data['test-AUC-mean'].max()]
```

```python
n_iters = cv_data[cv_data['test-AUC-mean'] == cv_data['test-AUC-mean'].max()]['iterations'].values[0]
```

```python
n_iters
```

**Обучение на всех данных**

```python
params = {'cat_features': cat_features,
          'eval_metric': 'AUC',
          'loss_function': 'Logloss',
          'learning_rate': 0.11,
          'random_seed': 77,
          'verbose':100,
          'scale_pos_weight': 74/26,
          'iterations': 820}
```

```python
model = CatBoostClassifier(**params)
```

```python
model.fit(train_data)
```

**Предсказание на тесте**

```python
test['y_pred'] = model.predict_proba(test_data)[:,1]
```

```python
roc_auc_score(test['outflow'],test['y_pred'])
```

**Необходимая точность достигнута, дальнейшее моделирование нецелесообразно**
